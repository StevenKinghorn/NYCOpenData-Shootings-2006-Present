- Got weather data from https://www.ncei.noaa.gov/access/past-weather/New%20York
- weather readings taken from JFK international airport, NY US (USW00094789)
- Dataset: "ncei_noaa_weather_JFK_international_airport.csv"

```python
import pandas as pd

weather = "../data/ncei_noaa_weather_JFK_international_airport.csv" 
df_weather = pd.read_csv(weather, skiprows=1) # first row contains location data 
df_weather.head()
```

```python
df_weather.info()
```

- Need to check all dates are in format yyyy/mm/dd before converting them into datetime
- Truncate dataset to only include dates from 2006-01-01 to 2026-03-31

### df_weather KEY
 - Date - yyyy-mm-dd 
 - TAVG - Average temperature (Degrees Fahrenheit)
 - TMAX - Max temperature for the day (Degrees Fahrenheit)
 - TMIN - Min temperature for the day (Degrees Fahrenheit)
 - PRCP - Total Precipitation (Inches)
 - SNOW - Total Snowfall (Inches)
 - SNWD - Total Snow Depth (Inches)

```python
#Are all dates in the form of yyyy-mm-dd?

mask = df_weather["Date"].str.match(r"^\d{4}-\d{2}-\d{2}$", na=False)
invalid_dates = df_weather.loc[~mask, "Date"]
display(invalid_dates)
```

```python
print(f"Number of null or empty Date variables: {df_weather["Date"].isna().sum()}")
```

```python
#Let's convert date column into datetime
df_weather["Date"] = pd.to_datetime(df_weather["Date"])
df_weather.info()
```

```python
#Only need weather between 2006-01-01 and 2026-03-31
df_weather_2006_2026 = df_weather[ 
    (df_weather['Date'] >= '2006-01-01') &
    (df_weather['Date'] <= '2026-03-31')
    ].copy() 
```

```python
#checking for missing values
df_weather_2006_2026.isna().sum()
```

```python
#Where are the missing rows for Snow Depth?
#If in summer, probably doesn't matter that they are missing - can assume they're 0
df_weather_2006_2026[df_weather_2006_2026['SNWD (Inches)'].isna()]
```

- It's doubtful it snowed in August or even April, but let's check 7 days prior each of the three dates here to see if SNOW (Inches) has anything above 0.0
- And temperatures are well above freezing!

```python
df_weather_2006_2026[df_weather_2006_2026["Date"].between('2025-04-15', '2025-04-22')] 
```

```python
df_weather_2006_2026[df_weather_2006_2026["Date"].between('2025-08-05', '2025-08-12')] 
```

- Yes, Values are going to be zero for snow depth
- Replacing NaN with 0.0

```python
# Replacing missing snow depth values with zero due to the temperature and the weather conditions in previous 7 days
df_weather_2006_2026["SNWD (Inches)"] = df_weather_2006_2026["SNWD (Inches)"].fillna(0.0)
```

```python
#checking for missing values
df_weather_2006_2026.isna().sum()
```

```python
#Average temperature is completely missing so will remove column
df_weather_2006_2026 = df_weather_2006_2026.drop(columns=["TAVG (Degrees Fahrenheit)"])
```
